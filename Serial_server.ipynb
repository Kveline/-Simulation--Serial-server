```python
import random
import statistics

# customer object
class customer: 
    current_server = 'no-server'
    total_time_spent = 0
    def __init__(self, id, classified): 
        self.id = id
        self.classified = classified

    def joining_server(self, server):
        self.current_server = server

# server object
class server:
   # how many current customer in server 
   current_customers = 0;
   # list customer in server   
   list_customers = [];
   def __init__(self, name): 
        self.name = name
   def add_customers(self):
        self.current_customers += 1
   def servicing(self, time_range_start, time_range_end):
        list_service_time = [];
        print('\n')
        print('[SERVICING SERVER ', self.name, '. . .]')
        print('\n')
        for customer in self.list_customers:
            # cutomer data 
            print('Customer Id   : ', customer.id)
            print('Classified    : ', customer.classified)
            # servicing time
            time = random.randint(time_range_start, time_range_end)/10
            list_service_time.append(time)
            print('Service time  : ', time, ' Minutes')
            customer.total_time_spent += time
            print('\n')
        
        if(len(self.list_customers) > 0):
          print('Avarage service time in Server ', self.name , ' is ', float("{:.2f}".format(statistics.mean(list_service_time))), ' Minutes')

# creating list       
list_customers = [] 
server_A = server('A')
server_B = server('B')


def add_customer(index):
  # classified as type 1 or 2
  if random.randint(0,100) < 60:
    list_customers.append(customer(index, 1))
  else :
    list_customers.append(customer(index, 2))
  
  # will join server A or not ?
  m = server_A.current_customers

  # joining probability
  if(random.randint(0,100) < (1/(m+1)*100)):
    # join server A
    server_A.add_customers()
    list_customers[index].joining_server('A')


# MAIN CODE
# ------------------
# get total cutomer
print('How many customer? ')
total_customer = input()

# add customers
for i in range(0, int(total_customer)):
    add_customer(i)

# customer in server A
server_A.list_customers = [customer for customer in list_customers if customer.current_server == 'A']
server_A.servicing(1, 16)

# customer with class 2 are moving from server A  to B
server_B.list_customers = [customer for customer in server_A.list_customers if customer.classified == 2]
server_B.servicing(1, 12)

# Summary
print('\n[SUMMARY] \n')
# How many case of balks during simulation.
customer_balk = [customer for customer in list_customers if customer.current_server == 'no-server']
print('Case of balks : ', len(customer_balk), ' customers')
# each customer total time spent
customer_non_balk = server_A.list_customers + server_B.list_customers
print('\nEach customer total time spent : \n')

for customer in customer_non_balk :
    # cutomer data 
    print('Customer Id   : ', customer.id)
    print('Classified    : ', customer.classified)
    print('Time Spent    : ', "{:.2f}".format(customer.total_time_spent), 'Minutes \n')

```
